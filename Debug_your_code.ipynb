## Methodology : Isolate the code!

> For debugging a given problem, the favorable situation is when the problem is isolated in a small number of lines of code, outside of framework or application code, with a short cycle of modify, launch, fail.

1. **Fail the code reliably**: Find a test case that fails the code every time.
2. **Divide and Conquer**: Once you have a failing test case, isolate the faulty code.
     * Which module?
     * What function?
     * Which line of code?
     => Isolate a small reproducible error: a test case.
  
3. **Change only one thing** each time and rerun the failed test case.
4. **Use variable explorer**, console, debugger to figure out what is wrong.
5. **Take notes and be patient**, it may take a while !

> Once you have done this step, isolate a small piece of code reproducing the bug and fix it using that piece of code, add that code to your test suite.

## Basic principle: the unit test

In a simplified way, the unit test consists in **testing the program step by step**, by evaluating whether the output element of the program corresponds to what should be applied to the input element.  
Example: we consider an input element x, we apply a function to this element that we store in a new variable y.

```python
x = 4
```

```python
print(x)
```

```python
y = x / 2 + 0.5
```

```python
print(y)
```

Here the unit test is to check that our input element x matches what we want, then check that the function applied to x matches what we expect.  
When the program is richer and more complex, the same principle of unit testing applies: the inputs and outputs are checked step by step, for each line of the program.

## The unit test within a loop

In a loop, the unit test is also applicable.  
Consider the following program, `my_list` is a vector of characters on which we will apply a transformation: put in uppercase with the `.upper ()` function, the result will be stored in a variable named `result`.

```python
my_list = ['a', 'b', 'd', 'e']
print(my_list)
```

```python
for i in my_list:
    result = i.upper()
    print("Result is", result, " when my_list =", i)
```

Let's imagine that things go wrong (which sometimes happens!).  
Here we modify the entry element `my_list`: we want to add the missing letter` c` at the 3rd position.  
An input error (which can sometimes happen!) leads to add the number 3.  
*The `.insert ()` function is used here to insert an element at a defined position.*

```python
my_list.insert(2, 3) # We insert 3 (instead of 'c') at the 3 index position
print(my_list)
```

We now re-run the same loop as before ...

```python
for i in my_list:
    result = i.upper()
```

... but now our program is failing and we don't know why!  
We can read the returned error ...  
`AttributeError: 'int' object has no attribute 'upper'`  
... but the interpretation is not obvious when we begin in programming.  
With more experience, we may more easily understand the printed error and make a first diagnostic.

One proposed method is to go and inspect when the program has failed.  
To do this, let's evaluate for which value of `i` the program was interrupted.

```python
print(i)
```

So the program stopped when i takes the value 3.  
We then discover our error since we were not expecting an integer in our vector my_list.

For our understanding, we are also looking for what is the last result of the vector `result`.

```python
print(result)
```

The last stored and valid result is `B`, the program then stopped and could not execute for the next element.

## The unit test within a function

In a function, the unit test is also applicable.  
Consider the following function, `my_function` that transforms in uppercase the input.

```python
def my_function(list):
  result = [i.upper() for i in list]
  return(result)
```

```python
my_list = ['a', 'b', 'd', 'e']
my_function(my_list)
```

Let's imagine that things go wrong (which sometimes happens!).  
Here we modify the entry element `my_list`: we want to add the missing letter` c` at the 3rd position.  
An input error (which can sometimes happen!) leads to add the number 3.  
*The `.insert ()` function is used here to insert an element at a defined position.*

```python
my_list.insert(2, 3) # We insert 3 (instead of 'c') at the 3 index position
print(my_list)
```

We now re-run the same loop as before ...

```python
my_function(my_list)
```

... but now our program is failing again and we (suppose we) don't know why!  
We can read the returned error ...  
`AttributeError: 'int' object has no attribute 'upper'`  
... but the interpretation is not obvious when we begin in programming.  

Same tips are still available for functions.  
1. Inspect when the program has failed and evaluate for which value of `i` the program was interrupted.
2. Have a look at the last result.

But an important thing is variables and results are **encapsulated** within the function.  
In other words variables and results are **not available in the global environnement**.  
Debugging is then possible by inspecting **without execution of the function**. And now tips given above are actionable.

## Conclusions

In terms of debugging, many good practices are essential:

- Try to understand and interpret the printed error.

- Evaluate the inputs and outputs, confirm that the output corresponds to the expected.

- Execute the script in small parts, step-by-step, launch the full execution only when you have tested all the code.

- Use keyboard shortcuts for executions, which facilitates and encourages good practices.

- With the Spyder IDE for Python :
  - Ctrl + Enter: execution of the selection (part of the line, the entire line, several lines, etc.).
  - F9: execution of the selection or of the line if no active selection.
  - F5: file execution (full script).

- Use the Debugging Mode, the Variable Explorer and the Console to evaluate step-by-step both input and output elements.

- The principle of unit testing also applies to a loop, a function, a module : evaluate element after element and identify the point of dysfunction.
